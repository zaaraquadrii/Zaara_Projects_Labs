# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

*Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.*

```python
##########################
##########################

## TASK-6 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################


def plot_KNN_error_rate(xdata,ydata):
  error_rate = []
  test_scores = []
  train_scores = []

  X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.3, random_state=101)

  for i in range(1,40):
      knn = KNeighborsClassifier(n_neighbors=i)
      knn.fit(X_train, y_train)
      pred_i = knn.predict(X_test)

      error_rate.append(np.mean(pred_i != y_test))
      train_scores.append(knn.score(X_train,y_train))
      test_scores.append(knn.score(X_test,y_test))

  plt.figure(figsize=(12,8))
  plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
          markerfacecolor='red', markersize=10)
  plt.title('Error Rate vs. K Value')
  plt.xlabel('K')
  plt.ylabel('Error Rate')
  print()
  max_train_score = max(train_scores)
  train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
  print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))
  print()
  max_test_score = max(test_scores)
  test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
  print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))

  return test_scores
```

```python
unchanged_test_scores = plot_KNN_error_rate(unchanged_data,diabetes_data['Outcome'])
```

## Standardize the Variables
Standardization (also called z-score normalization) is the process of putting different variables on the same scale. Standardization transforms your data such that the resulting distribution has a mean of 0 and a standard deviation of 1.

$$ Z = {X - \mu \over \sigma}$$


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(diabetes_data.drop('Outcome',axis=1))
```

```python
scaled_data = scaler.transform(diabetes_data.drop('Outcome',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_data,columns=diabetes_data.columns[:-1])
df_feat.head()
```

```python
scaled_test_scores = plot_KNN_error_rate(scaled_data,diabetes_data['Outcome'])
```

## TASK-8: Plot the voronoi diagram for the TASK-7

```python
from matplotlib.colors import ListedColormap

# Select two features for the Voronoi diagram (e.g., Glucose and BMI)
X_voronoi = scaled_data_minmax[:, [1, 5]] # Glucose is at index 1, BMI is at index 5
y_voronoi = diabetes_data['Outcome']

# Train a KNN model on the selected features
knn_voronoi = KNeighborsClassifier(n_neighbors=5) # Using a reasonable number of neighbors
knn_voronoi.fit(X_voronoi, y_voronoi)

# Create a meshgrid
x_min, x_max = X_voronoi[:, 0].min() - 0.1, X_voronoi[:, 0].max() + 0.1
y_min, y_max = X_voronoi[:, 1].min() - 0.1, X_voronoi[:, 1].max() + 0.1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predict class for each point in the meshgrid
Z = knn_voronoi.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the Voronoi diagram
plt.figure(figsize=(10, 8))
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])
plt.contourf(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
plt.scatter(X_voronoi[:, 0], X_voronoi[:, 1], c=y_voronoi, cmap=cmap_bold,
            edgecolor='k', s=20)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("Voronoi Diagram of KNN Classification (Glucose vs. BMI)")
plt.xlabel("Glucose (MinMax Scaled)")
plt.ylabel("BMI (MinMax Scaled)")
plt.show()
```

## TASK-9: Use K-Fold cross validation on all the above classification experiments and present an analysis of the results you obtain.

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
knn_unscaled = KNeighborsClassifier(n_neighbors=20)
unscaled_scores = cross_val_score(knn_unscaled, unchanged_data, diabetes_data['Outcome'], cv=10)
```

```python
mean_unscaled_accuracy = np.mean(unscaled_scores)
print(f"Mean cross-validation accuracy for unscaled data: {mean_unscaled_accuracy:.4f}")
```

```python
knn_scaled_minmax = KNeighborsClassifier(n_neighbors=5)
scaled_scores_minmax = cross_val_score(knn_scaled_minmax, scaled_data_minmax, diabetes_data['Outcome'], cv=10)
```

```python
mean_scaled_minmax_accuracy = np.mean(scaled_scores_minmax)
print(f"Mean cross-validation accuracy for MinMax scaled data: {mean_scaled_minmax_accuracy:.4f}")
```

```python
print("\nAnalysis of K-Fold Cross-Validation Results:")
if mean_scaled_minmax_accuracy > mean_unscaled_accuracy:
    print("The MinMax scaled data resulted in a higher mean cross-validation accuracy compared to the unscaled data.")
    print("This suggests that MinMax scaling improved the performance of the KNN model on this dataset.")
elif mean_scaled_minmax_accuracy < mean_unscaled_accuracy:
    print("The unscaled data resulted in a higher mean cross-validation accuracy compared to the MinMax scaled data.")
    print("This suggests that MinMax scaling did not improve the performance of the KNN model on this dataset.")
else:
    print("The mean cross-validation accuracy is the same for both unscaled and MinMax scaled data.")
```
